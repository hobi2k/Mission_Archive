```python
# 라이브러리 가져오기
import numpy as np
import matplotlib.pyplot as plt
from dataclasses import dataclass
import evaluate

# 파이토치 관련 라이브러리
import torch
from torch.utils.data import DataLoader
from torch.nn.utils.rnn import pad_sequence
from torch.optim import AdamW
from torch.optim.lr_scheduler import LambdaLR
from tqdm.auto import tqdm

# 음악 관련 라이브러리
import librosa
import librosa.display
from datasets import Audio, load_dataset, DatasetDict
from functools import partial
from transformers import pipeline, Seq2SeqTrainer, Seq2SeqTrainingArguments
from transformers import WhisperProcessor, WhisperForConditionalGeneration
from transformers.models.whisper.english_normalizer import BasicTextNormalizer

# 허깅페이스 업로드
from huggingface_hub import notebook_login

# 음악 듣기용
from IPython.display import Audio as Au

# 그라디오 라이브러리
import gradio as gr
```

```python
# 상수 변수 정의
class Config:
    # 디바이스 변수
    DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
        
    # 시드값
    SEED = 42
```

```python
# 데이터셋 가져오기
minds = load_dataset("PolyAI/minds14", "en-US")
```

```python
# 데이터셋 구조 확인
minds
```

```python
# 전처리기 초기화
wprocessor = WhisperProcessor.from_pretrained(
    "openai/whisper-tiny", task="transcribe"
)
```

```python
# 리샘플링
sampling_rate = wprocessor.feature_extractor.sampling_rate
minds["train"] = minds["train"].cast_column("audio", Audio(sampling_rate=sampling_rate))
```

```python
# 전처리 함수
def preprocessing(instance):
    audio = instance["audio"]
    
    instance = wprocessor(
        audio=audio["array"],
        sampling_rate=audio["sampling_rate"],
        text=instance["english_transcription"]
    )
    
    instance["input_length"] = len(audio["array"]) / audio["sampling_rate"]
    
    return instance
```

```python
# 전처리 실행
processed_dataset = minds.map(
    preprocessing, remove_columns=minds.column_names["train"], num_proc=1
)
```

```python
processed_dataset
```

```python
# 음성 길이 필터링
max_input_length = 30.0

def audio_filtering(length):
    return length < max_input_length
```

```python
# 필터 적용
processed_dataset["train"] = processed_dataset["train"].filter(
    audio_filtering,
    input_columns=["input_length"]
)
```

```python
processed_dataset
```

```python
# 데이터 분할
# 슬라이싱을 하면 리스트가 되어 버린다.
train_raw = processed_dataset["train"].select(range(450))
val_raw = processed_dataset["train"].select(range(450, len(processed_dataset["train"])))

processed_dataset = DatasetDict({
    "train":train_raw,
    "val":val_raw
})
```

```python
# 데이터 분할 확인
processed_dataset
```

```python
# 커스텀 Data Collator 정의, 가변적 샘플 길이에 패딩을 적용하기 위해 필요
@dataclass
class CustomCollator:
    processor: any
    
    def __call__(self, features: list) -> dict[str, torch.Tensor]:
        # 데이터셋 전처리 단계에서 "input_features"의 shape가 (batch, 1, T, D)
        input_features = [{"input_features": feature["input_features"][0]} for feature in features]
        # pad 추가 및 텐서 변환
        batch = self.processor.feature_extractor.pad(input_features, return_tensors="pt")
    
        label_features = [{"input_ids":feature["labels"]} for feature in features]
        labels_batch = self.processor.tokenizer.pad(label_features, return_tensors="pt")
        
        labels = labels_batch["input_ids"].masked_fill(labels_batch.attention_mask.ne(1), -100)
        
        # Whisper 계열 tokenizer는 텍스트 앞에 <|startoftranscript|> 같은 BOS 토큰을 붙인다.
        if (labels[:, 0] == self.processor.tokenizer.bos_token_id).all().cpu().item():
            labels = labels[:, 1:]
        
        batch["labels"] = labels
        
        return batch
```

```python
# data_collator 인스턴스화
data_collator = CustomCollator(processor=wprocessor)
```

```python
# wer 지표 인스턴스화
metric = evaluate.load("wer")
```

```python
# 텍스트 정규화 인스턴스화
normalizer = BasicTextNormalizer()
```

```python
# 정규화 wer 지표
def compute_metrics(pred, processor=wprocessor):
    pred_ids = pred.predictions
    label_ids = pred.label_ids
    
    label_ids[label_ids == -100] = processor.tokenizer.pad_token_id    
    
    pred_str = processor.batch_decode(pred_ids, skip_special_tokens=True)
    label_str = processor.batch_decode(label_ids, skip_special_tokens=True)
    
    wer_ortho = metric.compute(predictions=pred_str, references=label_str)
    
    pred_str_norm = [normalizer(pred) for pred in pred_str]
    label_str_norm = [normalizer(label) for label in label_str]

    pred_str_norm = [
        pred_str_norm[i] for i in range(len(pred_str_norm)) if len(label_str_norm[i]) > 0
    ]
    label_str_norm = [
        label_str_norm[i]
        for i in range(len(label_str_norm))
        if len(label_str_norm[i]) > 0
    ]

    wer = metric.compute(predictions=pred_str_norm, references=label_str_norm)
    
    return {"wer_ortho":wer_ortho, "wer":wer}
```

```python
# DataLoader 생성
train_loader = DataLoader(
    processed_dataset["train"],
    batch_size=15,
    shuffle=True,
    collate_fn=data_collator,
    drop_last=True
)

val_loader = DataLoader(
    processed_dataset["val"],
    batch_size=15,
    shuffle=False,
    collate_fn=data_collator,
    drop_last=True
)
```

```python
# 모델 인스턴스화
model = WhisperForConditionalGeneration.from_pretrained("openai/whisper-tiny").to(Config.DEVICE)
```

```python
# 모델 설정
model.config.use_cache = False
model.config.forced_decoder_ids = None
```

```python
optimizer = AdamW(model.parameters(), lr=1e-5)
```

```python
# 간단한 Warmup 스케줄러
def lr_lambda(step):
    warmup_steps = 50
    if step < warmup_steps:
        return float(step) / float(max(1, warmup_steps))
    return 1.0

scheduler = LambdaLR(optimizer, lr_lambda)
```

```python
# 훈련 루프
def train_one_epoch(model, loader, optimizer, scheduler):
    model.train()
    total_loss = 0.0

    pbar = tqdm(loader, desc="Training")

    for batch in pbar:
        batch = {k: v.to(Config.DEVICE) for k, v in batch.items()}

        outputs = model(
            input_features=batch["input_features"],
            labels=batch["labels"]
        )

        loss = outputs.loss
        loss.backward()

        optimizer.step()
        scheduler.step()
        optimizer.zero_grad()

        total_loss += loss.item()
        pbar.set_postfix({"loss": total_loss / (len(pbar) + 1)})

    return total_loss / len(loader)
```

```python
# 검증 루프
def validate(model, loader, processor):
    model.eval()
    total_wer = []

    with torch.no_grad():
        for batch in tqdm(loader, desc="Validation"):
            batch = {k: v.to(Config.DEVICE) for k, v in batch.items()}

            pred_ids = model.generate(
                input_features=batch["input_features"],
                max_length=225
            )

            # decode
            pred_str = processor.batch_decode(pred_ids, skip_special_tokens=True)

            labels = batch["labels"].clone()
            labels[labels == -100] = processor.tokenizer.pad_token_id
            label_str = processor.batch_decode(labels, skip_special_tokens=True)

            # compute WER
            wer_score = metric.compute(predictions=pred_str, references=label_str)
            total_wer.append(wer_score)

    return np.mean(total_wer)
```

```python
# 에포크 설정
best_wer = 999
EPOCHS = 30

for epoch in range(EPOCHS):
    print(f"\n===== Epoch {epoch+1} / {EPOCHS} =====")

    train_loss = train_one_epoch(model, train_loader, optimizer, scheduler)
    print(f"Train Loss: {train_loss:.4f}")

    val_wer = validate(model, val_loader, wprocessor)
    print(f"Validation WER: {val_wer:.4f}")

    # 베스트 모델 저장
    if val_wer < best_wer:
        best_wer = val_wer
        torch.save(model.state_dict(), "my_whisper_tiny.pth")
        print(">>> Best model saved.")

print("Training finished.")
print(f"BEST WER = {best_wer:.4f}")
```
